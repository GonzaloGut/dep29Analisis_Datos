```python
# Importar las bibliotecas necesarias
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.functions import broadcast
from pyspark.sql.types import *
```

```python
# Crear una instancia de SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# 1. Crear un DataFrame con datos de personas, que incluya nombre, edad y ciudad de residencia.

data = [
    ("Ana", 23, "New York"),
    ("Carlos", 35, "Los Angeles"),
    ("María", 29, "Chicago"),
    ("José", 41, "New York"),
    ("Lucía", 27, "Houston"),
    ("Pedro", 33, "Phoenix"),
    ("Sofía", 21, "San Diego"),
    ("Luis", 38, "New York"),
    ("Valeria", 26, "Los Angeles"),
    ("Miguel", 45, "Dallas"),
    ("Daniela", 31, "Houston"),
    ("Jorge", 28, "Chicago"),
    ("Camila", 24, "New York"),
    ("Andrés", 36, "San Antonio"),
    ("Paola", 34, "Phoenix"),
    ("Fernando", 42, "San Jose"),
    ("Rosa", 39, "New York"),
    ("Diego", 22, "Austin"),
    ("Natalia", 25, "Houston"),
    ("Raúl", 47, "Los Angeles"),
    ("Andrea", 30, "San Diego"),
    ("Héctor", 51, "New York"),
    ("Karen", 28, "Seattle"),
    ("Oscar", 44, "Miami"),
    ("Patricia", 37, "Chicago"),
    ("Alonso", 19, "New York"),
    ("Verónica", 33, "Denver"),
    ("Ricardo", 40, "Phoenix"),
    ("Elena", 26, "Houston"),
    ("Manuel", 48, "Los Angeles")
]

df = spark.createDataFrame(data,["Nombre","Edad","Ciudad"])

df.show(10)
```

```python
# 2. Mostrar solo los nombres de las personas del DataFrame.
df.select("Nombre").show(10)
```

```python
# 3. Filtrar personas cuya edad sea mayor o igual a 25.
df.filter(df["Edad"] >= 25).show(10)
```

```python
# 4. Agregar una nueva columna "Pais" con un valor constante para todas las filas.
df.withColumn("País", lit("Estados Unidos")).show(10)
```

```python
# 5. Calcular el promedio de edad de todas las personas.
df1 = df.groupBy().agg(avg("Edad").alias("Promedio de edades")).show()
```

```python
# 6. Ordenar el DataFrame por edad en orden descendente.
df.orderBy("Edad",ascending = False).show(10)
```

```python
# 7. Agrupar por ciudad y calcular la cantidad de personas en cada ciudad.
df.groupBy("ciudad").agg(count("Nombre").alias("Cantidad de personas x ciudad")).show()
```

```python
# 8. Renombrar la columna "Nombre" a "NombreCompleto".
df.withColumnRenamed("Nombre","NombreCompleto").show(10)
```

```python
# 9. Eliminar la columna "Edad" del DataFrame.
df.drop("Edad").show(10)
```

```python
# 10. Realizar una consulta SQL en el DataFrame para seleccionar personas mayores de 20 años.
df.filter(df["Edad"] > 20).show(10)
```

```python
# 11. Calcular la suma total de todas las edades.
df.agg(sum("Edad").alias("Suma total de edades")).show(10)
```

```python
# 12. Calcular la edad mínima y máxima de todas las personas.
df.select(min("Edad").alias("Edad mínima"),max("Edad").alias("Edad máxima")).show()
```

```python
# 13. Filtrar personas cuya ciudad de residencia sea "Chicago" y edad sea menor de 30.
df.filter(df["ciudad"] == "Chicago").show()
```

```python
# 14. Agregar una nueva columna "EdadDuplicada" que contenga el doble de la edad.
df.withColumn("Edad duplicada", df["Edad"]*2).show(10)
```

```python
# 15. Convertir todas las edades en años a meses.
df.withColumn("Edad en meses",df["Edad"]*12).show(10)
```

```python
# 16. Contar el número total de personas en el DataFrame.
df.groupBy().count().show()
```

```python
# 17. Filtrar personas cuya edad sea un número par.
df.where(df["Edad"] % 2 == 0).show(10)
```

```python
# 18. Calcular la cantidad de personas por rango de edades (0-20, 21-40, 41-60, 61+).
df.withColumn("Rango de edades",
              when(df["Edad"]<=20,"0-20")
              .when(df["Edad"]<=40,"21-40")
              .when(df["Edad"]<=60,"41-60")
              .when(df["Edad"]>=61,"61+")
              ).groupBy("Rango de edades" ).count().show()
```

```python
# 19. Contar cuántas personas tienen el mismo nombre.
df.groupBy(df["Nombre"]).count().show(10)
```

```python
# 20. Concatenar las columnas "Nombre" y "Ciudad" en una nueva columna llamada "InformacionPersonal".
df.withColumn("InformacionPersonal",concat(df["Nombre"],df["Ciudad"])).show(10)
```
